```python
# Iterate images and adapt them for training with tensorflow 
# By cropping, resizing and padding
import os
import tensorflow as tf
import pandas as pd
from PIL import Image
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array, array_to_img

def crop_image(image, boxes):
    width, height = image.size
    left, upper, right, lower = int(boxes["XMin"] * width), int(boxes["YMin"] * height), int(boxes["XMax"] * width), int(boxes["YMax"] * height)
    image_cropped = image.crop((left, upper, right, lower))
    return image_cropped

def save_image_with_postfix(image, filename, postfix="_1"):
    while os.path.exists(filename):
        base, extension = os.path.splitext(filename)
        filename = f"{base}{postfix}{extension}"
        # Increment the postfix (e.g., "_1", "_2", "_3", ...)
        postfix = f"_{int(postfix[1:]) + 1}"

    image.save(filename)

def resize_and_pad_image(image_pillow, target_size=(100, 100), padding_color=255):
    # Load the image
    #img = image.load_img(image_path)

    # Convert the image to a NumPy array
    img_array = img_to_array(image_pillow)

    # Get the original image dimensions
    original_size = img_array.shape[:2]

    # Calculate the resize ratio
    resize_ratio = min(target_size[0] / original_size[0], target_size[1] / original_size[1])

    # Resize the image
    resized_img = tf.image.resize(img_array, size=(int(original_size[0] * resize_ratio), int(original_size[1] * resize_ratio)))

    # Pad the image
    pad_x = (target_size[0] - resized_img.shape[0]) // 2
    pad_y = (target_size[1] - resized_img.shape[1]) // 2
    padded_img = tf.image.pad_to_bounding_box(resized_img, pad_x, pad_y, target_size[0], target_size[1])

    # Convert the NumPy array back to an image
    padded_img = array_to_img(padded_img.numpy())

    return padded_img

def crop_resize_and_pad_images(classes_to_prepare, image_types, src_dir_images_template, src_dir_boxes_template, dst_dir_images_template):
    """
    Stupid

    Parameters:
        classes_to_prepare: some description
        image_types: some description
        src_dir_template: some description
        dst_dir_template: some description
    """
    for _, class_description in classes_to_prepare.iterrows():
        for image_type in image_types:
            input_folder_images=src_dir_images_template.replace("{prefix}", image_type).replace("{class_displayname}", class_description["DisplayName"])
            input_folder_boxes=src_dir_boxes_template.replace("{prefix}", image_type).replace("{class_displayname}", class_description["DisplayName"])
            input_file_boxes=f'{class_description["DisplayName"]}_boxes.csv'
            full_path_boxes=os.path.join(input_folder_boxes, input_file_boxes)
            output_folder_images= dst_dir_images_template.replace("{prefix}", image_type).replace("{class_displayname}", class_description["DisplayName"])

            os.makedirs(output_folder_images, exist_ok=True)
            
            for filename in os.listdir(output_folder_images):
                if os.path.isfile(os.path.join(output_folder_images, filename)):
                    os.remove(os.path.join(output_folder_images, filename))

            boxes_for_class=pd.read_csv(full_path_boxes, dtype={'ImageID': str})

            for _, row in boxes_for_class.iterrows():
                image_id=row["ImageID"]

                full_path_image=input_folder_images + str(image_id) + ".jpg"
                original_image = Image.open(full_path_image)
                cropped_image = crop_image(original_image, row)
                original_image.close()
                processed_image = resize_and_pad_image(cropped_image)
                try:
                    # TODO implement saving rgba too
                    output_path = os.path.join(output_folder_images, image_id + '.jpg')  # Adjust the extension if needed
                    save_image_with_postfix(processed_image, output_path)
                except:
                    print(f"error saving {output_path}, continuing...")

input_folders_template = "./images/original/{prefix}/{class_displayname}/"
input_folders_boxes = "./generated_metadata/boxes/{prefix}/"
output_folders_template = "./images/processed/{prefix}/{class_displayname}/"
prefixes = ["train", "test"]

crop_resize_and_pad_images(
    classes_to_prepare=class_descriptions_subset,
    image_types=prefixes,
    src_dir_images_template=input_folders_template,
    src_dir_boxes_template=input_folders_boxes,
    dst_dir_images_template=output_folders_template)
```

```python
# Create dataset for training, validation and testing

import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from sklearn.model_selection import train_test_split
import numpy as np
import os

def load_images_and_labels(directory):
    images = []
    labels = []
    
    class_names = sorted(os.listdir(directory))
    label_dict = {class_name: idx for idx, class_name in enumerate(class_names)}
    
    for class_name in class_names:
        class_directory = os.path.join(directory, class_name)
        for filename in os.listdir(class_directory):
            filepath = os.path.join(class_directory, filename)
            img = load_img(filepath, target_size=(100, 100))
            img_array = img_to_array(img)
            images.append(img_array)
            labels.append(label_dict[class_name])
    
    return images, labels

# Load train and test images and labels
train_images, train_labels = load_images_and_labels(train_images_directory)
test_images, test_labels = load_images_and_labels(test_images_directory)

# Split train set into train+validation
train_images, validation_images, train_labels, validation_labels = train_test_split(
    train_images, train_labels, test_size=0.2, random_state=42
)

# Convert lists to numpy-arrays
train_images = np.array(train_images)
train_labels = np.array(train_labels)
validation_images = np.array(validation_images)
validation_labels = np.array(validation_labels)
test_images = np.array(test_images)
test_labels = np.array(test_labels)

# Normalize from 0-255 to 0-1
train_images, validation_images, test_images = train_images / 255.0, validation_images / 255.0, test_images / 255.0

# Print shapes
print("Train shapes:", train_images.shape, train_labels.shape)
print("Validation shapes:", validation_images.shape, validation_labels.shape)
print("Test shapes:", test_images.shape, test_labels.shape)


# Create TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_labels))
validation_dataset = tf.data.Dataset.from_tensor_slices((validation_images, validation_labels))
test_dataset = tf.data.Dataset.from_tensor_slices((test_images, test_labels))

# Print dataset element shapes
print("Train dataset element shapes:", train_dataset.element_spec)
print("Validation dataset element shapes:", validation_dataset.element_spec)
print("Test dataset element shapes:", test_dataset.element_spec)

# Shuffle trainining set and batch
batch_size = 32
train_dataset = train_dataset.shuffle(len(train_images)).batch(batch_size)
validation_dataset = validation_dataset.batch(batch_size)
test_dataset = test_dataset.batch(batch_size)
```
